```python
import pandas as pd
import altair as alt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("heart_2020_cleaned.csv")
```

```python
def convert_to_int(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
df['HeartDisease'] = df['HeartDisease'].apply(lambda x: convert_to_int(x))
```

```python
df.head()
```

```python
table_1 = df[['AgeCategory', 'Sex', 'Race', 'HeartDisease']].groupby(['AgeCategory', 'Sex', 'Race']).count()
table_1 = table_1.rename(columns={'HeartDisease': 'total_num_people'})
```

```python
table_2 = df[['AgeCategory', 'Sex', 'Race', 'HeartDisease']].groupby(['AgeCategory', 'Sex', 'Race']).sum()
table_2 = table_2.rename(columns={'HeartDisease': 'num_heart_disease'})
```

```python
age_group = pd.merge(table_1, table_2, on=['AgeCategory', 'Sex', 'Race'])
```

```python
age_group['percentage'] = round(age_group['num_heart_disease'] / age_group['total_num_people'] * 100, 1)
```

```python
age_group = age_group.reset_index()
```

```python
age_group[15:25]
```

```python
age_sex_race = alt.Chart(age_group, title='Distribution of heart disease among different sex/races'
            ).mark_bar(
            height=25).encode(
            x=alt.X('percentage', title='Percentage of Heart Disease'),
            y=alt.Y('AgeCategory'),
            color='Race',
            column=alt.Column('Sex', title=None),
            tooltip=['Sex:N', 'Race:N', 'percentage:Q']
            ).properties(
            width=350,
            height=400)
```
